# linear algebra


```python
import numpy as np  
```

```python
A = np.array([
    [2, -1],
    [0, 3]
])      

B = np.array([
    [1, 2],
    [3, -4]
])      

B = np.array([
    [4, 1],
    [-2, 1]
])    
```

```python
print("A + B is ", A+B)
```

```python
print("A*B=", A@B)
print("A*B (using np.dot)=", np.dot(A, B))
```

```python
print("3A - 2B=", 3*A - 2*B)
```

```python
print("determinant of A=", np.linalg.det(A))
```

```python
# print(A.shape)
rows, cols= A.shape
print(f"Matrix A has {rows} rows and {cols} columns.")
detA=np.linalg.det(A)

if rows != cols:
    print("Matrix A is not square, so it is not invertible.")

if detA != 0:
    print("Matrix A is invertible.")

Ainv= np.linalg.inv(A)   #matrix inverse=(1/detA) * adj(A)
print("Inverse of A is:\n", Ainv)

adjA= detA * Ainv
print("Adjoint of A is:\n", adjA)
```

```python
eigenval, eigenveec= np.linalg.eig(A)
print("eigenvalues of A =", eigenval)
print(eigenval[0])
print(eigenval[1])
print("eigenvectors of A =\n", eigenveec)
```

```python
x=np.array([
    [1,4,1.5,100],
    [1,6,2.0,150]
]
)
y=np.array([
    [20],
    [30]
])
a=x.T@x
b=x.T@y
beta=np.linalg.pinv(a)@b
print("Coefficients (beta) are:\n", beta)
print(y)
```

```python
car3 = np.array([[1, 8, 3.0, 200]])
price_car3 = car3 @ beta
print("Price for Car 3:", price_car3[0, 0])
```
