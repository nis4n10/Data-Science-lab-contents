## DATA VISUALIZATION


```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df = pd.read_csv("Health_heart_experimental.csv")
df.head(11)
```

```python
df_sorted = df.sort_values(by='age', ascending=False)
sns.lineplot(x='age',y='SysBP' ,data=df_sorted, color='GREEN', marker='o')
```

 Looking at the above line plot ,it is seen that the normal systolic blood pressure is 120 mmHG.On observing the data clearly it is seen that the BP has gone upto 126.5 mmhg and it lies under the age group of 60-65.the cause of this high blood pressue in this age group may be due to stress, work pressure, or lifestyle factors.

 likewise,talking about the lowest BP it is in between 124-125mmHG and age group of 30-35 are seen to be having the lowest BP .The cause of this can be due to  younger adults often have more active routines or less stress related to family and responsibilities.

 Overall, the graph shows that people between 2060 years tend to have higher blood pressure, maybe due to the combined effects of aging, work, and also due to  family responsibilities.

```python

age_groups = df.groupby('age')['BMI'].mean().reset_index()
sns.lineplot(x='age', y='BMI', data=age_groups, color='red', marker='o')
```

As we know,the normal BMI range is 18.524.9, but looking at the graph, almost every age group has a higher BMI.It means almost every age group has overweight or obesity . The 2030 age group are seen  to be age group with highest BMI, which means the young generation of  today are at more risk for heart problems and mental health issues.

A big reason for this is unhealthy eating habits. Children, adults are seen often having   fast food and other high-calorie foods and high preservatives containing food as snacks which is too bad for health.  Also adult people are busy in  their office life due to which they are seen having meals from restaurants or fast-food places which isn't very healthy.

Overall, across all ages people dont get enough time and also don't give enough time  for exercise or self-care, which led to the problem of higher BMI and obesity.

```python
# Create age groups
bins = [0, 20, 30, 40, 50, 60, 100]
labels = ['0-20','21-30','31-40','41-50','51-60','61+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Average BP by age group
avg_bp = df.groupby('age_group')[['SysBP','DiaBP']].mean().reset_index()
plt.ylabel("SysBP/DiaBP")
# Plot
sns.barplot(x='age_group', y='SysBP', data=avg_bp, color='yellow', label='Systolic')
sns.barplot(x='age_group', y='DiaBP', data=avg_bp, color='black', label='Diastolic')
# sns.barplot(x='age_group', y='SysBP', hue='DiaBP', data=avg_bp)
```

As discussed above every age group are at a equal risk of different heart diseases and other problems due to the higher BP.But mostly the new generation are at higher risk due to their boring lifestyle using mobile phones all the time and not giving enough time to their health.So,everybody should add exercises,yoga to their personal routine for happy and health life.

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
df = pd.read_csv("Health_heart_experimental.csv")

df['sex'].value_counts().plot(kind='pie', autopct='%0.3f', label='Male : 0/Female : 1',colors=['Yellow', 'green'])
```

This pie chart shows about the numbers of male and females that do checkups about their BP and according to this pie-chart we came to know that the number is almost similar .However ,the female are seen more concerned about their health than men.

```python

sns.boxplot(data=df, x='indication', y='BMI', hue='sex',palette={0: 'red', 1: 'blue'})
```

The boxplot shows a clear difference in BMI based on the Indication label.  Indication 1 people have very higher BMI and median around 35 and median is 25 for Indication 0 people. This means that individuals in the high-risk group are generally more overweight.

Gender doesnt seem to make a difference here because males and females have almost same BMI distributions which means that both sexes face similar risks for obesity-related issues.

Interestingly, some people in the Indication 1 group have very low BMI values i.e 1317, which appear as outliers. This indicates that the high-risk category may include individuals who are  underweight or have some medical conditions.The Indication 0 group doesnt have  outliers.

Overall, the graph shows that most people in the high-risk group have  higher BMI, making them more prone to heart disease and mental stress, while the gender does not affect this pattern.
