## DATA VISUALIZATION


```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df = pd.read_csv("Health_heart_experimental.csv")
df.head(11)
```

```python
df_sorted = df.sort_values(by='age', ascending=False)
sns.lineplot(x='age',y='SysBP' ,data=df_sorted, color='GREEN', marker='o')
```

 Looking at the above line plot ,it is seen that the normal systolic blood pressure is 120 mmHG.On observing the data clearly it is seen that the BP has gone upto 126.5 mmhg and it lies under the age group of 60-65.the cause of this high blood pressue in this age group may be due to stress, work pressure, or lifestyle factors.

 likewise,talking about the lowest BP it is in between 124-125mmHG and age group of 30-35 are seen to be having the lowest BP .The cause of this can be due to  younger adults often have more active routines or less stress related to family and responsibilities.

 Overall, the graph shows that people between 2060 years tend to have higher blood pressure, maybe due to the combined effects of aging, work, and also due to  family responsibilities.
